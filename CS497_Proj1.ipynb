```python
from sklearn.model_selection import train_test_split
```

```python
only_ints = one_hot_df.select_dtypes(include=['float64','int64','uint8'])
only_ints = only_ints.apply(lambda x: x.fillna(x.mean()))
only_ints = pd.concat([only_ints, one_hot_df['LotArea_decile']],
                      axis=1)
X_train, X_test, y_train, y_test = train_test_split(only_ints.drop('SalePrice',axis=1),
                                                    only_ints['SalePrice'],
                                                    test_size=0.2,
                                                    random_state=42)
```

6) **Let’s say you want to buidl a simple rules-based model that can be used as a baseline by find the decile of square feet that a home falls into and “predicting” the average price for that decile. Implement this logic using X_train from above and build a ‘model’ to find y_train and y_test based on X_train and X_test.** 

```python
avg_price_decile = {x: 0 for x in range(0,10)}
for x, val in enumerate(prelim_df.groupby('LotArea_decile')['SalePrice'].mean()) :
    avg_price_decile[x] = val
```

```python
avg_price_decile
```

```python
X_train['yHat'] = X_train['LotArea_decile'].apply(lambda x: avg_price_decile[x-1])
```

```python
X_test['yHat'] = X_test['LotArea_decile'].apply(lambda x: avg_price_decile[x-1])
```

7) **Build a function to find the mean-squared error taking in as argument vectors y_pred and y_actual. Use the function to find the mean-squared error on your training and test sets.**

```python
def MSE(y_actual, y_pred) :
    try:
        y_actual = y_actual.to_numpy()
    except:
        pass
    try:
        y_pred = y_pred.to_numpy()
    except:
        pass
    diff = (y_actual - y_pred) ** 2
    return diff.sum() / len(y_actual)
```

```python
print('MSE for train set :', MSE(y_train, X_train['yHat']))
print('MSE for test set :', MSE(y_test, X_test['yHat']))
```

9) **Use scikit-learn linear model LinearRegression to run a model through your data. Evaluate the model on your x_test to get predicted y_pred. Compare y_pred to y_test to find MSE. Does the linear regression model beat your original rules-based model?**
