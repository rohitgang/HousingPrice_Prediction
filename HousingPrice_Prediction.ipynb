Creating an arbitrary model using Multiple Linear Regression Model

```python
X_Features= X[['MSSubClass','LotFrontage','LotArea','OverallQual','OverallCond','YearBuilt','YearRemodAdd','MasVnrArea','HalfBath','BedroomAbvGr','KitchenAbvGr','TotRmsAbvGrd','Fireplaces','GarageYrBlt','GarageCars','GarageArea','WoodDeckSF','OpenPorchSF','EnclosedPorch','PoolArea','MiscVal','MoSold','YrSold']]
```

```python
X_Features.fillna(X_Features)
X_Features= np.asanyarray(X_Features.values)
X_Features.shape
```

```python
import math

for i in range(1460):
    for j in range(23):
        arr= X_Features[i][j]
        if math.isnan(arr) :
            X_Features[i][j]= 0
```

```python
from sklearn.model_selection import train_test_split

Y= np.asanyarray(df['SalePrice'].values)
X_train, X_test, Y_train, Y_test= train_test_split(X_Features, Y, test_size= 0.3, random_state= 4)
```

```python
from sklearn.linear_model import LinearRegression

model= LinearRegression()
model.fit(X_train, Y_train)
```

```python
yHat= model.predict(X_test)
```

Evaluating the model

```python
def rmsle(y_pred, y_test) : 
    assert len(y_test) == len(y_pred)
    return np.sqrt(np.mean((np.log(1+y_pred) - np.log(1+y_test))**2))
```

```python
from sklearn import metrics

print('MSE:', (np.mean(Y_test - yHat)**2))
print('R2 score:', metrics.r2_score(yHat, Y_test))
print('RSMLE:',rmsle(np.absolute(yHat), np.absolute(Y_test)))
```

With multiple linear regression, rsmle of 0.21417 was achieved.
The model consisted of int and float values only. So a vast amount of features were neglected for the training of model.
